```python
import pandas as pd
import logging
from simpletransformers.language_modeling import LanguageModelingModel

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

file_path = '/content/ideas.csv'
data = pd.read_csv(file_path)
data.columns = ['Content']
data = data['Content'].unique()
```

```python
train_size = int(len(data) * 0.8)
train_data = data[:train_size]
test_data = data[train_size:]
len(train_data), len(test_data)
```

```python
!rm /content/train.txt /content/test.txt
```

```python
with open('/content/train.txt', 'a') as f:
    
    for pair in train_data:
        
        f.writelines(pair + '\n')
        
print('Train File Saved.')

with open('/content/test.txt', 'a') as f:
    
    for pair in test_data:
        
        f.writelines(pair + '\n')
        
print('Test File Saved.')
```

```python
train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "block_size": 512,
    "max_seq_length": 512,
    "learning_rate": 5e-6,
    "train_batch_size": 8,
    "gradient_accumulation_steps": 8,
    "num_train_epochs": 30,
    "mlm": False,
    "output_dir": "/content/fine_tuned_model",
}

model = LanguageModelingModel("gpt2", "gpt2", args=train_args)
model.train_model('/content/train.txt', eval_file = '/content/test.txt')
results = model.eval_model('/content/test.txt')
print(results)
```

```python
from simpletransformers.language_generation import LanguageGenerationModel

my_model = LanguageGenerationModel(
    "gpt2", "/content/fine_tuned_model", args={"max_length": 250}
)
```

```python
prediction_data = []

for i in range(10):

  predictions = my_model.generate('', verbose = False)
  predictions = predictions[0].split('.')

  for prediction in predictions:

    if prediction not in prediction_data:

      if str(prediction).startswith(' '):

        prediction = prediction[1:-1]

      prediction_data.append(prediction)

with open('test_predictions.txt', 'a') as f:

  for prediction in prediction_data:

    f.writelines(str(prediction) + '\n')

print('Done!!!')
```
