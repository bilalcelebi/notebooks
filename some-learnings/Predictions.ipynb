```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import numpy as np

train_path = '/home/bilalcelebi/Workspace/notebooks/data/regression/train.csv'
test_path = '/home/bilalcelebi/Workspace/notebooks/data/regression/test.csv'
```

```python
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
```

```python
train.shape, test.shape
```

```python
train_na_index = train.isna().sum()
train_na_index = train_na_index.loc[train_na_index.values > 0]
train_na_index = train_na_index.index.to_list()

for index in train_na_index:
    
    train.drop(str(index), axis = 1, inplace = True)
```

```python
test_na_index = test.isna().sum()
test_na_index = test_na_index.loc[test_na_index.values > 0]
test_na_index = test_na_index.index.to_list()

for index in test_na_index:
    
    test.drop(str(index), axis = 1, inplace = True)
```

```python
train = pd.get_dummies(train)
```

```python
cat_columns = []

for column in train.columns:
    if train[column].dtype == 'object':
        cat_columns.append(column)
```

```python
train.drop(cat_columns, axis = 1, inplace=True)
train
```

```python
X = train.drop('SalePrice', axis = 1)
y = train['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=125, test_size=0.2, shuffle=True)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
mes = mean_squared_error(preds, y_test)
np.sqrt(mes)
```
